```python
def initialize_parameters(layer_dims):

  np.random.seed(3)
  parameters = {}
  L = len(layer_dims)

  for l in range(1, L):
    parameters['W' + str(l)] = np.ones((layer_dims[l-1], layer_dims[l]))*0.1
    parameters['b' + str(l)] = np.zeros((layer_dims[l], 1))
  return parameters
```

```python
initialize_parameters([2,2,1])
```

```python
def linear_forward(A_prev,  W, b):
  Z = np.dot(W.T, A_prev) + b

  return Z
```

```python
# forward propagation

def forward_propagation(X, parameters):
  A = X
  L = len(parameters) // 2

  for l in range(1, L+1):
    A_prev = A
    W1 = parameters['W' + str(l)]
    b1 = parameters['b' + str(l)]

    #print("A" + str(l-1) + ": ", A_prev)
    #print("W" + str(l)+ ": ", W1)
    #print("b" + str(l) + ": ", b1)



    A = linear_forward(A_prev, W1, b1)
    #print("A" + str(l) + ": ", A)


  return A, A_prev
```

```python
def backward_propagation(parameters, y, y_pred, A1, X, l_rate):
  parameters['W2'][0][0] = parameters['W2'][0][0] + (l_rate * 2 * (y - y_pred) * A1[0][0])
  parameters['W2'][1][0] = parameters['W2'][1][0] + (l_rate * 2 * (y - y_pred) * A1[1][0])
  parameters['b2'][0][0] = parameters['b2'][0][0] + (l_rate * 2 * (y - y_pred))


  parameters['W1'][0][0] = parameters['W1'][0][0] + (l_rate * 2 * (y - y_pred) * parameters['W2'][0][0] * X[0][0])
  parameters['W1'][0][1] = parameters['W1'][0][1] + (l_rate * 2 * (y - y_pred) * parameters['W2'][0][0] * X[1][0])
  parameters['b1'][0][0] = parameters['b1'][0][0] + (l_rate * 2 * (y - y_pred) * parameters['W2'][0][0])

  parameters['W1'][1][0] = parameters['W1'][1][0] + (l_rate * 2 * (y - y_pred) * parameters['W2'][1][0] * X[0][0])
  parameters['W1'][1][1] = parameters['W1'][1][1] + (l_rate * 2 * (y - y_pred) * parameters['W2'][1][0] * X[1][0])
  parameters['b1'][1][0] = parameters['b1'][1][0] + (l_rate * 2 * (y - y_pred) * parameters['W2'][1][0])
```

```python
# now implementing epoch
parameters = initialize_parameters([2,2,1])

epochs = 24

for i in range(epochs):
  loss = []
  for j in range(df.shape[0]):
    X = df[['cgpa', 'profile_score']].values[j].reshape(2,1)
    y = df[['LPA']].values[j][0]


    y_pred, A1 = forward_propagation(X, parameters)
    y_pred = y_pred[0][0]

    backward_propagation(parameters, y, y_pred, A1, X, l_rate=0.001)
    loss.append(loss_function(y_pred, y))

  print('Epoch - ', i+1, 'Loss - ', np.array(loss).mean())
```
